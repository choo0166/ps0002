#### Using dplyr verbs (select, filter, arrange, mutate, group_by and summarise)

```r
# Selects a set of columns
select(stats, name, sleep_total)
```

```r
# Select all the columns except specified column range
select(stats, -(awake:brainwt))
```

```r
# Select a range of columns
select(stats, name:sleep_rem)
```

```r
# Select all columns that start with a specified string
# Additional options include contains() - Select columns that contain a character string,
# matches() - Select columns that match a regular expression,
# one_of() - Select columns names that are from a group of names
select(stats, starts_with('sl'))  # Use ends_with to specify ending string
```

```r
# Filter the rows for mammals that sleep a total of more 
# than 16 hours and have a body weight of greater than 1 kilogram
filter(stats, sleep_total >= 16, bodywt >= 1)
```

```r
# Filter the rows for mammals with order as “Perissodactyla” and “Primates”
filter(stats, order %in% c('Perissodactyla', 'Primates'))
```

```r
# Using pipe operator to select two columns (name and sleep_total) and then 
# show the first 6 rows (head) of the selected data frame
stats %>% select(name, sleep_total) %>% head()
```

```r
# Select three columns from sleepdata, arrange the rows by the order and then arrange the rows by sleep_total. 
# Finally filter the rows for mammals that sleep for 16 or more hours
# Note arrange sorts columns in ascending order unless using desc()
stats %>% select(name, sleep_total, sleep_rem, sleep_cycle) %>% arrange(sleep_cycle, desc(sleep_total)) %>% filter(sleep_total >= 16)
```

```r
# Grouping with group_by() allows us to break down the dataframe
# into common row entries and summarize in terms of certain variables
# Example: Grouping by order and summarise count of each order
stats %>% group_by(order) %>% summarise(Count = n())  # n() counts number of rows
```

```r
# More examples of grouping and summarise
# Argument na.rm=TRUE tells R to remove all missing values which are recorded as “NA” in the dataset
order_sleep_stats <- stats %>% group_by(order) %>% summarise(sleep_avg = mean(sleep_total, na.rm=T))
order_sleep_stats
```

```r
# Creating new columns using mutate(data, variable1 =..., variable2 =...,)
df <- stats %>% select(name, sleep_total, sleep_rem, sleep_cycle) %>% mutate(nonrem_sleep= sleep_total - sleep_rem)
df
```

```r
# Creating new columns using mutate (with ifelse condition)
mutate(df, sleep_ge5 = ifelse(nonrem_sleep > 5, 'Yes', 'No'))
```

#### Lab 3 Task

```r
sleep2 <- stats %>% select(vore:bodywt) %>% filter(brainwt < 0.012)
sleep2
```

```r
sleep2 <- sleep2 %>% mutate(rem_percentage = sleep_rem * 100/sleep_total, cycles = sleep_total/sleep_cycle)
sleep2
```

```r
summary(sleep2)
```

```r
summary <- sleep2 %>% filter(!is.na(vore)) %>% group_by(vore) %>% summarise(Count = n(), rem_percentage_avg = mean(rem_percentage, na.rm=T), cycles_avg = mean(cycles, na.rm=T), bodywt_avg = mean(bodywt, na.rm=T))
summary
```

```r
# Perform a comparison of the continuous variable over different categories with boxplots
sleep2_plot <- sleep2 %>% filter(!is.na(vore)) %>% select(vore, rem_percentage) %>% group_by(vore)
ggplot(sleep2_plot, aes(x=vore, y=rem_percentage)) + geom_boxplot()
```

```r
sleep2_plot0 <- sleep2 %>% filter(!is.na(vore)) %>% select(rem_percentage, cycles)
ggplot(sleep2_plot0, aes(x=rem_percentage, y=cycles)) + geom_point()
```

```r
ggplot(sleep2_plot0, aes(x=rem_percentage, y=cycles)) + geom_point() + geom_smooth(method='lm')
```
