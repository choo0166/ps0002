```r
library(ggplot2)
```

```r
stats <- read.csv('msleep_ggplot2.csv', header=T);stats
```

```r
# Define summary statistic function that returns the range (max-min) of a variable
rg<-function(x){
    x <- na.omit(x)
    max(x)-min(x)
}
```

```r
# Q1
sleep_stats <- stats %>% filter(!is.na(vore)) %>% group_by(vore) %>% summarise(count=n(), avg_sleep=mean(sleep_total, na.rm=T), sd_sleep=sd(sleep_total, na.rm=T), range_sleep=rg(sleep_total));sleep_stats
```

```r
stats %>% subset(vore=='insecti') %>% select(vore, sleep_total)
stats %>% subset(vore=='carni') %>% select(vore, sleep_total)
```

```r
# Q2
# Perform a comparison of the continuous variable over different categories
# Note that geom_bar produces a categorical histogram (bar chart)
# Argument 'stat' specifies the statistical transformation used on the data in this layer,
# stat = 'bin' means the height of each bar equals to the number of cases in each group, 
# stat = 'identity' means the height of each bar represents the value in the each group.
ggplot(sleep_stats, aes(x=vore, y=avg_sleep)) + geom_bar(stat='identity')
ggplot(sleep_stats, aes(x=vore, y=sd_sleep)) + geom_bar(stat='identity')
ggplot(sleep_stats, aes(x=vore, y=range_sleep)) + geom_bar(stat='identity')

# To explain why vore group with largest sd differs from group with largest range
# Insecti group has the largest sd. This group includes only 5 observations, 
# which do not concentrate around the mean value, resulting in a large variance
# and hence sd. Note that the carni samples have outliers that lie on 
# two extremes resulting in the greater range.
```

```r
# Q3
# The histogram divides the x axis into bins (breaks, intervals), counts the number of observations 
# in each bin and display the counts with bars
# Syntax: geom_histogram(data = NULL, stat = "bin", position = "stack",   ..., bins = NULL, na.rm = FALSE, show.legend = NA, ...)
# Note: when imposing the densityplot onto the histogram, the vertical axis of 
# the histogram should be density, instead of count (default setting in hist) to make them comparable.
ggplot(stats, aes(sleep_total, fill=vore)) + geom_histogram(bins=30, na.rm=T)
ggplot(stats, aes(sleep_total)) + geom_histogram(aes(y=..density..), binwidth=0.8, colour='black', fill='white', na.rm=T) + geom_density(alpha=.2, fill='pink')

ggplot(stats, aes(sleep_rem, fill=vore)) + geom_histogram(bins=30, na.rm=T)
ggplot(stats, aes(sleep_rem)) + geom_histogram(aes(y=..density..), binwidth=0.8, colour='black', fill='white', na.rm=T) + geom_density(alpha=.2, fill='pink')
```

```r
# Q4,5
sleep1 <- stats %>% na.omit() %>% filter(brainwt < 0.1) %>% mutate(rem_ratio=sleep_rem/sleep_total); sleep1
sleep1 %>% select(rem_ratio, brainwt) %>% ggplot(aes(x=brainwt, y=rem_ratio)) + geom_point()

# Scatterplot relationship for groups of data
ggplot(sleep1, aes(x=brainwt, y=rem_ratio, col=vore)) + geom_point()

# Yes. For carni mammals, the relationship between both variables is close to a 
# linearlly increasing pattern. Carni mammals with heavier brainwt tend to have higher rem-ratio. 
# For other vore groups, the rem_ratio is almost unrelated to brainwt.

# Verify by calculating r-squared values
remlm <- lm(rem_ratio ~ brainwt, data=sleep1)
summary(remlm)$r.squared
# Focus on carni group
sleep2 <- sleep1 %>% filter(vore=='carni')
rem_carni_lm <- lm(rem_ratio ~ brainwt, data=sleep2)
summary(rem_carni_lm)$r.squared
```
