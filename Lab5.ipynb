#### Statistical Modelling

##### Definition of response and predictor variables
- The variable we would like to predict is the _response_ variable, denoted by $Y$ and the individual observations $y_{i}$.
- The variables that the response variable depend on are the _predictor_ or independent variables, denoted by $X = (X_{1}, X_{2},...,X_{p})$ and individual observations as $x_{ij}$, where $i$ indexes the observations 1,...,$n$, and $j$ indexes the predictor variables 1,...,$p$. [Think of observations as rows of a dataframe, and the variables as columns]

##### Statistical model
- Assume that the response variable $Y$ relates to the predictors $X$ by $Y = f(X) + \epsilon$, where $f$ is the unknown function that relates Y to X, and $\epsilon$ is the amount that $Y$ differs from $f(X)$.
- A statistical model is any algorithm that estimates $f$ and we denote the estimator function $\hat f$.

##### Prediction vs estimation
- Inference problems require a set of observations of predictors $(x_{i1}, x_{i2},...,x_{ip})$ to predict a value for the response variable, $\hat y_{i}$ and $\hat y_{i} = \hat f(x_{i1}, x_{i2},...,x_{ip})$. This requires $\hat f$ to be explicitly defined.
- In prediction problems, the form of $\hat f$ does not matter, instead it depends on minimizing $\epsilon$ such that the prediction $\hat y_{i}$ is as close to the observed value $y_{i}$ as possible.
- Some models are more suited for inference, and others for prediction.

##### Regression vs Classification
- There are 2 main types of prediction problems:
  - **Regression problems** are ones with a _quantitative_ response variable, e.g. predicting the number of    cab pick-ups in Singapore.
  - **Classification problems** are ones with a _categorical_ response variable, e.g. predicting whether or not a Netflix user will like a particular movie.
  
##### Loss function
- In order to quantify how well a model performs, we define a loss or error function.
- A common loss function for quantitative outcomes is the **Mean Squared Error (MSE)** $= \frac{1}{n}\sum\limits_{i=1}^n(y_{i} - \hat y_{i})^{2}$, used to calculate the _least squares_ regression line. 
  
##### Linear Regression estimation model
- Linear regression is a **parametric** model with a fixed number of parameters and independent of observed data.
- Aim is to fit a line to the data, using the _least squares_ regression line.
- Assume relationship between $Y$ and $X$ is given by a linear function, $Y = f(X) + \epsilon = \beta_{0} + \beta_{1}X + \epsilon$, where $\beta_{0}$ and $\beta_{1}$ are unknown parameters.
- We compute the estimated regression line, given by $\hat Y = \hat f(X) = \hat\beta_{0} + \hat\beta_{1}X$, where $\hat\beta_{0}$ and $\hat\beta_{1}$ are estimates of $\beta_{0}$ and $\beta_{1}$ respectively.
- Aim is to choose $\hat\beta_{0}$ and $\hat\beta_{1}$ such that $\epsilon$ is minimized, using available observations.
- Using MSE as the loss function, we can find explicit formulae for $\hat\beta_{0}$ and $\hat\beta_{1}$, i.e.
$\hat\beta_{1} = \frac{\sum\limits_{i} (x_{i}-\bar{x})(y_{i}-\bar{y})}{\sum\limits_{i} (x_{i}-\bar{x})^{2}}$, $\hat\beta_{0} = \bar{y}-\hat\beta_{1}\bar{x}$, where $\bar{x}$ and $\bar{y}$ are sample means.
- The line $\hat{Y} = \hat\beta_{0} + \hat\beta_{1}X$ is known as the least squares regression line of $Y$ on $X$.

##### k-Nearest Neighbours (kNN) prediction model
- kNN is a **non-parametric** model whose set of parameters depends entirely on the training set.
- Aim is to predict a response from a set of observed predictor values using the responses of other observations most similar or nearest to it.
- Unlike linear regression, kNN ‘regression’ does not return a single formula but rather a best predicted output value for any given input.
- The predicted response, $\hat y_{i}$ for the $i^{th}$ observation of a predictor variable, $x_{i}$ is the average of the observed responses, $y_{j}$ of the k-closest (in terms of distance between the predictors, $\lvert x_{i}-x_{j} \rvert$) observations, i.e. $\hat y_{i} = \frac{1}{k} \sum\limits_{j=1}^k y_{j}$.

##### k-Nearest Neighbours regression example

```r
# Load the data
data("Boston", package = "MASS")
# Inspect the data and summary
head(Boston); dim(Boston); summary(Boston)
```

```r
# Split data into training and test sets
# Use sample() to randomly draw a sample of the specified size from the dataset either 
# without replacement or with replacement (same index can appear more than once) and returns a vector of indices
# Syntax: sample(x, size, replace=F, prob=NULL)
# Use 80/20 splitting rule (80% of data for training, 20% for testing)
set.seed(100)  # Set a seed for random number generator to ensure reproducibility
training_idx <- sample(1: nrow(Boston), nrow(Boston)*0.8) 
train_data <- Boston[training_idx, ]; head(train_data)
test_data <- Boston[-training_idx, ]; head(test_data)
```

```r
library(caret) # Load R's ML library

# Use train() function to fit the model on the training set
# Syntax: train(formula, data, method='knn',...,trControl=trainControl(), preProcess=c('center', 'scale'), tuneLength=3)
# Note preProcess=c('center', 'scale') standardizes the data as the distance measure in kNN 
# is sensitive to the predictor's scale. See section 8.2.2 at https://bradleyboehmke.github.io/HOML/knn.html#eq:euclidean
# In general, the formula of a model is specified as y~x1+x2+..., where y is the outcome variable we are
# predicting and x1, x2,... are the predictor variables. tuneLength specifies the number of possible k values to evaluate
# and selects the best one among them.

# See https://kevinzakka.github.io/2016/07/13/k-nearest-neighbor/ on parameter tuning
set.seed(101)  # Set a seed to ensure reproducibility of train()
model <- train(
medv~., data = train_data, method = "knn",  # ~. means all other variables will be used as predictor variables
trControl = trainControl("cv", number = 10), # Note that cross-validation resamples and splits the training set by the specified number of times, choose a number that ensures size of resampled data set >= 50   
preProcess = c("center","scale"),
tuneLength = 10
)
plot(model)
```

##### Selecting a k value
- The best k is the one that minimize the prediction error RMSE (root mean squared error).
- RMSE $= \sqrt{\sum\limits_{i=1}^n\frac{(\hat y_{i} - y_{i})^{2}}{n}}$
- train() code above choses a value of k that minimizes the RMSE.


```r
# Best tuning parameter k that minimize the RMSE
model$bestTune  # best k=11
```

```r
# Make predictions on the test data using predict()
predictions <- predict(model, test_data); head(predictions)

# Compute the prediction error RMSE
RMSE(predictions, test_data$medv)

# To visualize the performance of kNN regression, plot predicted medv vs medv in test_data
plot(test_data$medv, predictions, main="Prediction performance of kNN regression")
```

##### Summary of kNN regression
- Remember to standardize the data when performing kNN analysis.
- Test multiple k-values to decide an optimal value for your data. This can be done automatically using train() in the caret package.
- Assess the model performance on test data.
- Advantages: Non-parametric algorithm without assumptions. Parametric models like linear regression has lots of assumptions to be met by data before it can be implemented. 
- Disadvantages: **No explicit model** (we are not able to interpret the effects of predictor variables on the outcome). Computationally intensive, since in order to predict an outcome for a new value (in the test set), it will need to search through all the points in its training set to find the nearest ones.

##### Linear regression example

```r
# lm() function is used to fit a linear regression model
# Syntax: lm(formula, data,...)

lmodel <- lm(medv~., data=train_data)
summary(lmodel)  # prints summary of model fitting
```

##### Interpreting the lm() summary
- R-squared value is a measure of fitting performance, the higher the R-squared (as close to $\lvert1\rvert$), the better fitted the model is. 
- R-squared $\approx 0.72$ implies that the linear model using all possible predictor variables in the data set can explain 72% of variation of the outcome, medv. 
- Values under Estimate are the estimated coefficients $\hat\beta$ in the linear regression model $Y = \beta_{0} + \beta_{1}X_{1} +...+ \beta_{p}X_{p} + \epsilon, p = 13$.
- The fitted model based on the training data is $\hat Y = \hat\beta_{0} + \hat\beta_{1}X_{1} +...+ \beta_{p}X_{p}$. 
- The important impacts of predictors for outcome medv are indicated by \* for its significance (by p-value). The more number of *’s, the greater the significance level of the predictor in the model.

```r
# Make predictions on the test data
predictions <- predict(lmodel, test_data); head(predictions)

# Compute the prediction error RMSE
RMSE(predictions, test_data$medv)

# Visualize the performance of linear regression
plot(test_data$medv, predictions, main="Prediction performance of linear regression")
```

RMSE of linear regression is not as good as that of kNN regression. In the case where relationship between outcome and predictors is nonlinear, kNN reg would be better than linear regression.

We can look at the residuals, or the difference between the observed value and the predicted value i.e. ($y_{i} - \hat y_{i}$) to assess if the linear regression model fits the training data well and whether the fitting measure R-squared can be improved.

```r
# Calculate residuals
residuals(lmodel)

# Create various residual plots
plot(lmodel)
 
```

We'll focus on the first graph that shows the residuals on the vertical axis and the fitted $\hat y$ on the horizontal axis (0). If the points are randomly dispersed around the horizontal axis, a linear regression model is appropriate for the data; otherwise, a non-linear model is more appropriate. 



In this case, it looks like it is possible to improve the model fitting by incorporating quadratic terms.

```r
# Prepare data
flights <- mutate(flights, gain=arr_delay-dep_delay)
flights_ua <- flights %>% filter(carrier=='UA', air_time > 550); head(flights_ua)
summary(flights_ua)

# Visualize the relationship
library(ggplot2)
flights_ua %>% select(gain, air_time) %>% ggplot(aes(x=air_time, y=gain)) + geom_point() + geom_smooth()
```

```r
# Perform kNN regression

# Split data into training and test sets
flights_ua <- flights_ua %>% select(air_time, gain); flights_ua

set.seed(100)  
trg_idx <- sample(nrow(flights_ua), nrow(flights_ua)*0.8) 
trg_data <- flights_ua[trg_idx, ]; head(trg_data)
test_data_0 <- flights_ua[-trg_idx, ]; head(test_data_0)
```

```r
# Fit model on training set

set.seed(101)  
flights_model_knn <- train(
gain~., data = trg_data, method = "knn",  
trControl = trainControl("cv", number = 10),  
preProcess = c("center","scale"),
tuneLength = 10
)
plot(flights_model_knn)

# Best tuning parameter k that minimize the RMSE
flights_model_knn$bestTune   # best k=15
```

```r
# Make predictions on the test data
flight_predictions <- predict(flights_model_knn, test_data_0); head(flight_predictions)

# Compute the prediction error RMSE
RMSE(flight_predictions, test_data_0$gain)

# To visualize the performance of kNN regression, plot predicted medv vs medv in test_data
plot(test_data_0$gain, flight_predictions, main="Prediction performance of kNN regression")
```

```r
# Perform linear regression

flights_lmodel <- lm(gain~., data=trg_data)
summary(flights_lmodel)
```

The fitted model is gain = $-504.4369 + 0.8093$air_time, with R-squared $\approx$ 0.60, so air_time is linearly correlated with gain

```r
# Make predictions on the test data
lm_predictions <- predict(flights_lmodel, test_data_0); head(lm_predictions)

# Compute the prediction error RMSE
RMSE(lm_predictions, test_data_0$gain)

# Visualize the performance of linear regression
plot(test_data_0$gain, lm_predictions, main="Prediction performance of linear regression")
plot(x=test_data_0$air_time, y=test_data_0$gain)
abline(flights_lmodel, col='red')  # plot regression line
```

From the RMSE and prediction plot, the linear regression model performs better than knn regression.
